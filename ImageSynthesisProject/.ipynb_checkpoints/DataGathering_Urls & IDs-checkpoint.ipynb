```python
# Must have imports
from IPython.display import display, HTML
from bs4 import BeautifulSoup
import requests

# Creating an http session using the requests module
IDR_BASE_URL = "https://idr.openmicroscopy.org"
INDEX_PAGE = "%s/webclient/?experimenter=-1" % IDR_BASE_URL
# create http session
with requests.Session() as session:
    request = requests.Request('GET', INDEX_PAGE)
    prepped = session.prepare_request(request)
    response = session.send(prepped)
    if response.status_code != 200:
        response.raise_for_status()
```

```python
def to_url(starting_id, ending_id):
    links = []
    IMAGE_DETAILS_URL = "{base}/webclient/imgData/{image_id}/"
    for IMAGE_ID in range(starting_id, ending_id + 1):
        qs = {'base': IDR_BASE_URL, 'image_id': IMAGE_ID}
        url = IMAGE_DETAILS_URL.format(**qs)
        r = session.get(url)
        RENDER_IMAGE = "{base}/webgateway/render_image/{image_id}/0/0/"
        img_url = RENDER_IMAGE.format(**qs)
        links.append(img_url)
    return links
```

```python
URL = []
for elems1, elems2 in IDs:
    print('Current URLs being generated are in the range: ',(elems1, elems2))
    URL.append(to_url(elems1, elems2))
```

```python
def download_imgs(url, ids, file_format='.jpg'):
    """
    args:
        url: the url from which to download the 
        images from
        ids: the image ids
        file_format: the format of the file to be 
        saved to, .jpg by default
    
    return:
        this function does not return anything
    
    """
    response = requests.get(url)
    file_name = f'{ids}{file_format}'
    file = open(f'{file_name}', 'wb')
    file.write(response.content)
    file.close
    
    return None
```
